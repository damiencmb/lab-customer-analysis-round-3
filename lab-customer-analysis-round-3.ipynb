```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('marketing_customer_analysis.csv')
df
```

```python
#Show DataFrame info
df.info()
```

```python
#Describe DataFrame
df.describe(include='all').T
```

```python
response = df['Response'].value_counts()
response.plot(kind='bar')
plt.show()
```

```python
#Show a plot of the response rate by the sales channel.
group=df.groupby(['Sales Channel'])
group['Response'].value_counts().unstack().plot(kind='bar')
```

```python
group_per=group['Response'].value_counts('Yes').unstack()*100
group_per.plot(kind='bar')
```

```python
# Define the bin edges for total claim amount
bin_edges = [0, 250, 500, 750, 1000, 1250, 1500, float('inf')]

# Create labels for the bins
bin_labels = ['0-250','251-500', '501-750','751-1000', '1001-1250', '1251-1500', '1500+']

# Cut the 'Total Claim Amount' into bins and create a new column 'Total Claim Amount Bracket'
df['Total Claim Amount Bracket'] = pd.cut(df['Total Claim Amount'], bins=bin_edges, labels=bin_labels)

# Group by 'Total Claim Amount Bracket' and calculate the response rate (mean of 'Response')
response_rate = df.groupby('Total Claim Amount Bracket')['Response'].value_counts(normalize=True).unstack()*100

# Create a bar plot
ax = response_rate.plot(kind='bar', stacked = True)
plt.xlabel('Total Claim Amount Bracket')
plt.ylabel('Response Rate')
plt.title('Response Rate by Total Claim Amount Bracket')
plt.xticks(rotation=45)

ax.yaxis.set_major_formatter('{x:.0f}%')
plt.show()
```

```python
# Calculate response rate by income bracket

# Define the bin edges for income
bin_edges = [-1, 20000, 40000, 60000, 80000, 100000, float('inf')]

# Create labels for the bins
bin_labels = ['0-20k', '20k-40k', '40k-60k', '60k-80k', '80k-100k', '100k+']

# Cut the 'Income' into bins and create a new column 'Income Bracket'
df['Income Bracket'] = pd.cut(df['Income'], bins=bin_edges, labels=bin_labels)

# raw_data['Income Bracket'].value_counts()
df.head()
```

```python
df['Income Bracket'].value_counts()
```

```python
response_by_income = pd.crosstab(df['Income Bracket'], df['Response']) #, normalize='index'

# response_by_income = raw_data.groupby('Income Bracket')['Response'].value_counts()#.unstack() #normalize=True
response_by_income.plot(kind='bar', stacked=True)
plt.title('Response Rate by Income')
plt.xlabel('Income Bracket')
plt.ylabel('Response Rate')
plt.legend(title='Response', loc='upper right', labels=['No', 'Yes'])
plt.show()
```
